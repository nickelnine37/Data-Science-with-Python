# 2. Matplotlib plotting

Take a look at the code below. You will see a function that creates a sin wave of a given frequency and amplitude.

````python
# this array represents the time interval betwen 0 and 1, with 501 time points
time = numpy.linspace(0, 1, 501)

def sin_wave(freq, amp):
    
    # this function returns a sin wave numpy array, for a given frequency and amplitude
    
    return amp * numpy.sin(2 * numpy.pi * freq * time)
````

Take a look at the code below, it plots two sin waves of different frequency.

````python
plt.figure()

signal1 = sin_wave(freq=1, amp=3)
signal2 = sin_wave(freq=4, amp=1)

plt.plot(time, signal1)
plt.plot(time, signal2)
````

## a) Add a key

Cope the code aboe into the cell below. Add a label to both lines using `label='F={}, A={}}'` where the frequency and amplitude of the wave is indicated. Remember to use `plt.legend()` to make the label appear as a key on the plot. 

````python
plt.figure()

signal1 = sin_wave(freq=1, amp=3)
signal2 = sin_wave(freq=4, amp=1)

plt.plot(time, signal1, label='F=1, A=3')
plt.plot(time, signal2, label='F=4, A=1')

plt.legend()
````

## b) Add an x-label, a y-label and a title. 

Copy and paste your code from above into a new cell below. Use `plt.xlabel()` and `plt.ylabel()` to show that the x-axis measures time and the y-axis measures amplitude. Add an appropriate title using `plt.title()`. 

````python
plt.figure()

signal1 = sin_wave(freq=1, amp=3)
signal2 = sin_wave(freq=4, amp=1)

plt.plot(time, signal1, label='F=1, A=3')
plt.plot(time, signal2, label='F=4, A=1')

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Oscillations graph')

plt.legend()
````

# 3. Matplotlib scatter plots

In this question we use random numpy arrays to simulate weight and height measurements of a sample of 100 male and female people in the UK. Run the code below to simulate these measurements. 

````python
# Run this cell! Don't worry if you don't understand exactly what's going on
# Although we envourage you to take a look and see if you can 

sample_size = 100

male_height = numpy.random.normal(175, 7.4, sample_size)
female_height = numpy.random.normal(161, 7.1, sample_size)

male_weight = male_height * 0.6 + numpy.random.normal(-24, 7, sample_size)
female_weight = female_height * 0.5 + numpy.random.normal(-12, 5, sample_size)
````

## a) `mean()`

Use the function `numpy.mean()` to find the mean of all samples. 

````python
# your code here

print(numpy.mean(male_height))
print(numpy.mean(female_height))
print(numpy.mean(male_weight))
print(numpy.mean(female_weight))
````

## b) Scatter plot labels

Complete the code below by adding labels for the different groups (remember to use `label='whatever'` AND `plt.legend()`), as well as axis labels and a title.  

````python
plt.figure()

plt.scatter(male_height, male_weight, label='Males')
plt.scatter(female_height, female_weight, label='Females')

plt.legend()
````

## c) Add transparency

Copy and paste your code from above into the cell below. Add the `alpha` argument to make the points somewhat transparent 

````python
# your code here

plt.figure()

plt.scatter(male_height, male_weight, label='Males', alpha=0.5)
plt.scatter(female_height, female_weight, label='Females', alpha=0.5)

plt.legend()
````
