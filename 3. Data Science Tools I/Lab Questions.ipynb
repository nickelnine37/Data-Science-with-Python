# 2. Matplotlib plotting

Take a look at the code below. You will see a function that creates a sin wave of a given frequency and amplitude.

````python
# this array represents the time interval betwen 0 and 1, with 501 time points
time = numpy.linspace(0, 1, 501)

def sin_wave(freq, amp):
    
    # this function returns a sin wave numpy array, for a given frequency and amplitude
    
    return amp * numpy.sin(2 * numpy.pi * freq * time)
````

Take a look at the code below, it plots two sin waves of different frequency. Feel free to play around with it!

````python
plt.figure()

signal1 = sin_wave(freq=1, amp=3)
signal2 = sin_wave(freq=4, amp=1)

plt.plot(time, signal1)
plt.plot(time, signal2)
````
